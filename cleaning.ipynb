```python

locs = {}
def get_loc(loc):
    if not loc[0] in string.ascii_letters:
        loc = loc[5:]
    if loc in locs:
        return locs[loc]
    loc_url = loc.replace(',',' ').replace(' ','%20')
    response = requests.get('https://maps.googleapis.com/maps/api/geocode/json?address='+loc_url+'&key=AIzaSyDErx0jbn3Z9nkWw0zAod2ceJFqa8n7BiI')
    resp_json_payload = response.json()
    try:
        r= resp_json_payload['results'][0]['geometry']['location']
    except:
        print(loc,'not found')
        r= {'lat':0,'long':0}
    locs[loc] = r
    print(loc)
    print(locs)
    print()
    return r
def get_lat(loc):
    return get_loc(loc)['lat']
def get_lng(loc):
    return get_loc(loc)['lng']
    
get_loc('amsterdam')
```

```python
get_lat(df.loc[1,'adress'])
df['lat'] = df.adress.apply(get_lat)
df['long'] = df.adress.apply(get_lng)
```
